```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
cres_train_features = ["cres","tc","gdp","finance"]

cres_df = numerical_df[cres_train_features]
cres_df = cres_df.dropna()

xtrain,xtest,ytrain,ytest = train_test_split(cres_df[["tc","gdp","finance"]],cres_df["cres"])
print(xtrain.shape,ytrain.shape,xtest.shape,ytest.shape)
```

```python
cres_model = LinearRegression()
cres_model.fit(xtrain,ytrain)
```

```python
mean_squared_error(cres_model.predict(xtest),ytest)**(1/2)
```

```python
cres_missing_df = numerical_df[cres_train_features]
cres_missing_df = pd.merge(cres_missing_df,cres_missing_df[["tc","gdp","finance"]].dropna())
cres_missing_df = cres_missing_df[cres_missing_df['cres'].isnull()]
```

```python
x_missing = cres_missing_df[["tc","gdp","finance"]]
cres_model.predict(x_missing)
```

```python
cres_missing_df["cres"] = cres_model.predict(x_missing)
cres_missing_df.isnull().sum()
```
