```python
WASTE_GENERATION_df = pd.read_csv("C:/Users/hp/Desktop/Projects/Business-Sustainability-through-Environmental-Impact-Analysis/Data/Waste Generation/municipal waste.csv")
WASTE_GENERATION_df.head()
```

```python
categorical_columns = ["region","province","name","istat","alt","isle","sea","urb","fee","d_fee","sample","geo","s_wteregio","s_landfill"]
```

```python
WASTE_GENERATION_df.shape
```

```python
numerical_df = WASTE_GENERATION_df.drop(categorical_columns,axis=1)
numerical_df.head()
```

```python
numerical_df.isnull().sum()
```

```python
for col in numerical_df.columns:
    numerical_df[col].fillna(numerical_df[col].mean(),inplace=True)
```

```python
numerical_df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
round(numerical_df["tc"].sum()) #Total cost per captia spend in managing the waste
```

```python
round(numerical_df["cres"].sum()) #Total money saved
```

```python
round(numerical_df["pden"].mean()) # Average people per km2
```

```python
round(numerical_df["wden"].mean()) # Average Waste per km2
```

```python
plt.figure(figsize=(10,10))
data = numerical_df[["organic","paper","glass","wood","metal","plastic","raee","texile","other"]].describe().loc["mean"]
palette_color = sns.color_palette('bright')
plt.pie(data,labels = data.index,colors=palette_color,autopct='%.0f%%')
```
